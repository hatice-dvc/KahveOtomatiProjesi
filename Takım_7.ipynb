```python
import time

menu = {
    "1. Espresso":10.00,
    "2. Caffe latte":15.00,
    "3. Mocha":13.00,
    "4. Cappuccino":18.00,
    "5. Americano":11.00,
    "6. Türk Kahvesi":8.00,
    "7. Caffee Macchiato":17.00,
    "8. Filtre Kahve":12.00
}

menuAdedi = {
    "1. Espresso":0,
    "2. Caffe latte":0,
    "3. Mocha":0,
    "4. Cappuccino":0,
    "5. Americano":0,
    "6. Türk Kahvesi":0,
    "7. Caffee Macchiato":0,
    "8. Filtre Kahve":0
}

class Kahve_otomatı():

  def __init__(self,otomat_durum = "Kapalı",menu = {},menuAdedi = {}, kahve_listesi = []):
    self.otomat_durum = otomat_durum
    self.menu = menu
    self.kahve_listesi = kahve_listesi
    self.menu_adedi = menuAdedi

    
  def otomat_ac(self):

    if self.otomat_durum == "Açık":
      print("Kahve otomatı zaten açık!")
    else:
      print("Kahve otomatı açılıyor...")
      time.sleep(2)
      print("Kahve otomatı açıldı.")
      self.otomat_durum = "Açık"
  
  def otomat_kapat(self):

    if self.otomat_durum == "Kapalı":
      print("Kahve otomatı zaten kapalı!")
    else:
      print("Kahve otomatı kapatılıyor...")
      time.sleep(2)
      print("Kahve otomatı kapandı.")
      self.otomat_durum = "Kapalı"

  def display_menu(self):
    
    print("-----YETKAHVE MENU-----")

    for item in list(menu.keys()):
     print(f"{item.upper():<15} : {menu[item]} TL")
  
  def display_menuAdedi(self):

    print("-----YETKAHVE MENU ADET-----")

    for item in list(menuAdedi.keys()):
      print(f"{item.upper():<15} : {menuAdedi[item]} adet")
   
  def ekstra_kahve_ekle(self,kahve_ismi):

    print("Kahve ekleniyor..")
    time.sleep(1)

    self.kahve_listesi.append(kahve_ismi)
    print("Kahve eklendi.")
  
  def miktar_degistir(self,kahve_num,guncel_adet):
 
    if kahve_num == 1:
      menuAdedi["1. Espresso"] = guncel_adet
    if kahve_num == 2:
      menuAdedi["2. Caffe latte"] = guncel_adet
    if kahve_num == 3:
      menuAdedi["3. Mocha"] = guncel_adet
    if kahve_num == 4:
      menuAdedi["4. Cappuccino"] = guncel_adet
    if kahve_num == 5:
      menuAdedi["5. Americano"] = guncel_adet
    if kahve_num == 6:
      menuAdedi["6. Türk Kahvesi"] = guncel_adet
    if kahve_num == 7:
      menuAdedi["7. Caffee Macchiato"] = guncel_adet
    if kahve_num == 8:
      menuAdedi["8. Filtre Kahve"] = guncel_adet
    print(f"{kahve_num} numaralı kahvenin adedi {guncel_adet} olarak başarıyla güncellendi. ")
    kahve.display_menuAdedi()  
  
  def fiyat_guncelle(self, kahve_no, guncel_fiyat):

    if kahve_no == 1:
      menu["1. Espresso"] = guncel_fiyat
    if kahve_no == 2:
      menu["2. Caffe latte"] = guncel_fiyat
    if kahve_no == 3:
      menu["3. Mocha"] = guncel_fiyat
    if kahve_no == 4:
      menu["4. Cappuccino"] = guncel_fiyat
    if kahve_no == 5:
      menu["5. Americano"] = guncel_fiyat
    if kahve_no == 6:
      menu["6. Türk Kahvesi"] = guncel_fiyat
    if kahve_no == 7:
      menu["7. Caffee Macchiato"] = guncel_fiyat
    if kahve_no == 8:
      menu["8. Filtre Kahve"] = guncel_fiyat
    kahve.display_menu()
     
  def __str__(self):
    return "Eklenen kahveler: {}\n".format(self.kahve_listesi)

kahve = Kahve_otomatı()

print("""
        KAHVE OTOMATI UYGULAMASI
----Kahve otomatı uygulamasına hoşgeldiniz!!----
Lütfen yapmak istediğiniz işleminizi seçiniz:
1. Kahve otomatını açma
2. Kahve otomatını kapatma
3. Kahve menüsünü görme
4. Ekstra kahve ekleme
5. Eklenen kahveleri görüntüleme 
6. Kahve miktarını değiştirme
7. Kahve fiyatı güncelleme
Uygulamadan çıkmak istiyorsanız lütfen 'n' tuşuna basınız.
""")

while True:

  islem = input("Tercihiniz: ")

  if (islem == "n"):
    print("Uygulama Sonlandırıldı..")
    break
  
  elif (islem == "1"):
    kahve.otomat_ac()

  elif (islem == "2"):
    kahve.otomat_kapat()

  elif (islem == "3"):
    kahve.display_menu()
  
  elif (islem == "4"):
    kahve_isimleri = input("Ekstra eklemek istediğiniz kahve varsa lütfen , ile ayırarak giriniz: ")
    kahve_listesi = kahve_isimleri.split(",")
    for eklenecekler in kahve_listesi:
      kahve.ekstra_kahve_ekle(eklenecekler)

  elif (islem == "5"):
    print(kahve)

  elif (islem == "6"):
    kahve.display_menuAdedi()
    kahve_num = int(input("Lütfen ekleme yapmak istediğiniz kahvenin numarasını yazınız: "))
    guncel_adet = int(input("Kaç adet kahve istiyorsunuz: "))
    print("*"*30)
    kahve.miktar_degistir(kahve_num,guncel_adet)
    
  elif (islem == "7"):
    kahve.display_menu()
    kahve_no = int(input("Lütfen güncelleme yapmak istediğiniz kahvenin numarasını yazınız: "))
    guncel_fiyat = float(input("Kahvenin güncel fiyatını yazınız: "))
    print("*"*30)
    print("Kahvenin fiyatı başarıyla güncellendi!")
    kahve.fiyat_guncelle(kahve_no,guncel_fiyat)
    
  else:
    print("Geçersiz bir işlem yaptınız! Lütfen tekrar deneyiniz..")
```
